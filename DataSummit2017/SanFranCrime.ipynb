# Data Ingest

Here, the raw train data is read from csv flat file and saved as a more type-friendly parquet file

```python
import csv
import json
from dateutil import parser
from datetime import datetime
from pyspark.sql.types import *

# Replace these file paths with the locations for where you are working
sfcrime_raw_data = '/Users/bill.walrond/Documents/dsprj/data/SanFranCrime/train.csv'
sfcrime_schema_file = '/Users/bill.walrond/Documents/dsprj/data/SanFranCrime/train_schema.txt'
sfcrime_schema_json = '/Users/bill.walrond/Documents/dsprj/data/SanFranCrime/train_schema.json'

sfc_rdd = sc.textFile(sfcrime_raw_data)
    
print ('Line count of train file: %d' % sfc_rdd.count())

# Read and process the schema file
schemadict = {}
i = 0
schema = sc.textFile(sfcrime_schema_file)
line_cnt = schema.count()
print ('Lines in schema file: %d' % line_cnt)
for l in schema.collect():
    col = l.split(',')
    schemadict.update({i: (col[0],col[1])})
    i += 1
print ('Length of the schemadict: %d' % len(schemadict.keys()))

# Establish the Struct for the Schema
keys = schemadict.keys()
# keys.sort()
schema = StructType()
for k in sorted(keys):
    if schemadict[k][1] == 'int':
        schema.add(StructField(schemadict[k][0], IntegerType(), True))
    elif schemadict[k][1] == 'str':
        schema.add(StructField(schemadict[k][0], StringType(), True))
    elif schemadict[k][1] == 'float':
        schema.add(StructField(schemadict[k][0], FloatType(), True))
    elif schemadict[k][1] == 'datetime':
        schema.add(StructField(schemadict[k][0], TimestampType(), True))
    else:
        print ('Unsupported or incorrect data type')

print ('Writing schema json to file ...')
json_out = open(sfcrime_schema_json,'w')
json_out.write(json.dumps(schema.jsonValue(), indent=4, separators=(',', ': ')))
json_out.close()
        
# Skip the first header line with this trick ...
sfc_nh = sfc_rdd.zipWithIndex().filter(lambda tup: tup[1] > 0).map(lambda tup: tup[0])

def toRowSep(line):
    """Parses one row using csv reader"""
    for r in csv.reader([line], delimiter=','):
        return r

sfc_split = sfc_nh.map(toRowSep)
    
lens = sfc_split.map(lambda r: len(r))
print ('Max len: %d' % lens.max())
print ('Min len: %d' % lens.min())
sfc_split = sfc_split.cache()
```

```python
def convert_types(row,schema):
    d = row
    for col, data in enumerate(row):
        if col <= len(schema.keys()):
            typed = schema[col]
            if data is None:
                d[col] = None
            elif typed[1] == 'string':
                d[col] = data
            elif typed[1] == 'int':
                d[col] = int(round(float(data)))
            elif typed[1] == 'float':
                d[col] = float(data)
            elif typed[1] == 'datetime':
                # d[col] = data
                # d[col] = dateutil.parser.parse(data)
                d[col] = datetime.strptime(data,'%Y-%m-%d %H:%M:%S')
    return d

def toTypedRow(row):
    return convert_types(row, schemadict)

# Now, convert the types of all the rdd elements
sfc_typed = sfc_split.map(toTypedRow)
print(sfc_typed.take(1))
sfc_train = sqlContext.createDataFrame(sfc_typed, schema)
print("Dataframe row count: {:,}".format(sfc_train.count()))
```

```python
from pyspark.sql.functions import *
import pandas as pd

def profileDF(df):
    cols = [name for name,type in df.dtypes if type in ['int','bigint','float','smallint','tinyint','double'] ] # get a list of all the string columns
    summ_cols = ['Column','Null_cnt','Non-null_cnt','Pct_null','Unique_cnt','Max','Min','Avg','StdDev','Kurtosis','Skewness']
    summ_df = pd.DataFrame(columns=summ_cols)
    total_rows = sfc_train.count()
    for col in cols:
        sum_vals = []
        print('---- Summarizing: {0} ----'.format(col))
        nulls = df.filter(col+" is Null").count() 
        nonnulls = df.filter(col+" is not Null").count()
        sum_val = [col,
            nulls,       # count of nulls
            nonnulls,   # count of non-nulls
            nulls/float(total_rows),                        # pct of nulls
            df.select(col).distinct().count(),       # count of unique values
            df.agg({col:'max'}).first()[0],              # max
            df.agg({col:'min'}).first()[0],
            df.agg({col:'avg'}).first()[0],
            df.select(stddev(col)).first()[0],
            df.select(kurtosis(col)).first()[0],
            df.select(skewness(col)).first()[0]]             # min
        summ_df.loc[len(summ_df)] = sum_val
    
    summNums_df = sqlContext.createDataFrame(summ_df)
    
    
    cols = [name for name,type in df.dtypes if type == 'string' ] # get a list of all the string columns
    summ_cols = ['Column','Null_cnt','Non-null_cnt','Pct_null','Unique_cnt','Max_len','Min_len','Max_words','Min_words']
    summ_df = pd.DataFrame(columns=summ_cols)
    wordCnt = udf(lambda x: len(x.split()), IntegerType())
    total_rows = df.count()
    for col in cols:
        sum_vals = []
        print('---- Summarizing: {0} ----'.format(col))
        words = df.withColumn('wordcnt', wordCnt(col)).select('wordcnt')
        words = words.cache()
        nulls = df.filter(col+" is Null").count() 
        nonnulls = df.filter(col+" is not Null").count()
        sum_val = [col,
            nulls,       # count of nulls
            nonnulls,   # count of non-nulls
            nulls/float(total_rows),                        # pct of nulls
            df.select(col).distinct().count(),       # count of unique values
            df.select(max(length(col))).first()[0],  # max length
            df.select(min(length(col))).first()[0],  # min length
            words.agg({'wordcnt':'max'}).first()[0],                                    # max number of words
            words.agg({'wordcnt':'min'}).first()[0] ]                                    # min number of words
        summ_df.loc[len(summ_df)] = sum_val
    
    summStr_df = sqlContext.createDataFrame(summ_df)
    
    cols = [name for name,type in df.dtypes if type in ['timestamp','date'] ] # get a list of all the string columns
    summ_cols = ['Column','Null_cnt','Non-null_cnt','Pct_null','Unique_cnt','Latest','Earliest']
    summ_df = pd.DataFrame(columns=summ_cols)
    total_rows = df.count()
    for col in cols:
        sum_vals = []
        print('---- Summarizing: {0} ----'.format(col))
        nulls = df.filter(col+" is Null").count() 
        nonnulls = df.filter(col+" is not Null").count()
        sum_val = [col,
            nulls,       # count of nulls
            nonnulls,   # count of non-nulls
            nulls/float(total_rows),                        # pct of nulls
            df.select(col).distinct().count(),       # count of unique values
            df.select(max(col)).first()[0].strftime('%Y-%m-%d'),  # latest date
            df.select(min(col)).first()[0].strftime('%Y-%m-%d')   # earliest date
            ]                                
        summ_df.loc[len(summ_df)] = sum_val
    
    summDt_df = sqlContext.createDataFrame(summ_df)
   
    return summNums_df, summStr_df, summDt_df

df1, df2, df3 = profileDF(sfc_train)
df1.show(truncate=False)
df2.show(truncate=False)
df3.show(truncate=False)
```
